```python
import glob
import os
import sys

import numpy as np
import pandas as pd
from bokeh.charts import Line, show
from bokeh.io import output_notebook
from natsort import natsorted

output_notebook()

# Hard-code inclusion of Fabian's code
dirname = os.path.dirname(os.path.realpath('__file__'))
dirname = os.path.join(dirname, '../3rdparty/temporal_proposals/')
for i in ['datasets', 'evaluate']:
    sys.path.append(os.path.join(dirname, i))

from run_evaluation import evaluate_proposals
```

```python
tIOU = np.arange(0, 1, 0.05)
tIOU[0] = 0.01

dirname = '../data/experiments/thumos14/a00/'
prop_files = natsorted(glob.glob(os.path.join(dirname, '*.n-prop*')))
n_prop = np.zeros(len(prop_files))
recall = np.zeros((len(prop_files), tIOU.size))

for i, v in enumerate(prop_files):
    n_prop[i] = int(v.split('n-prop_')[1])
    for j, w in enumerate(tIOU):
        rst = evaluate_proposals('thumos14', v, w, 1.0)
        recall[i, j] = rst['recall']
```

```python
df = pd.DataFrame(recall.T)
df.columns = ['n-prop {}'.format(i) for i in n_prop]
df['tIOU'] = tIOU
p = Line(df, x='tIOU', y=df.columns[:-1].tolist(),
         ylabel='Recall', legend="top_right")
f = show(p)
```
